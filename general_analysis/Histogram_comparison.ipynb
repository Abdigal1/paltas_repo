```python
import numpy as np
import skimage
from skimage import io
import os
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
id=0
vu=0

SDB='../../Sentera_Data_Base/'
V=os.listdir(SDB)
NIR=os.path.join(SDB,V[vu],'NIR',os.listdir(NIR)[id])
RGB=os.path.join(SDB,V[vu],'RGB',os.listdir(RGB)[id])

inir=os.listdir(NIR)
irgb=os.listdir(RGB)

nir=io.imread(inir)
rgb=io.imread(irgb)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(15, 15))
rgbax = fig.add_subplot(1, 2, 1)
rgbax.imshow(nir)
plt.axis("off")
nirax = fig.add_subplot(1, 2, 2)
nirax.imshow(rgb)
plt.axis("off")
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(15, 15))
b = fig.add_subplot(3, 3, 1)
b.imshow(nir[:,:,0],cmap=plt.cm.gray)
plt.axis("off")
aux = fig.add_subplot(3, 3, 2)
aux.imshow(nir[:,:,1],cmap=plt.cm.gray)
plt.axis("off")
r = fig.add_subplot(3, 3, 3)
r.imshow(nir[:,:,2],cmap=plt.cm.gray)
plt.axis("off")

b_hist = fig.add_subplot(3, 3, 4)
b_hist.hist(nir[:,:,0].ravel(), bins=255)
plt.axis("off")

aux_hist = fig.add_subplot(3, 3, 5)
aux_hist.hist(nir[:,:,1].ravel(),bins=255)
plt.axis("off")

r_hist = fig.add_subplot(3, 3, 6)
r_hist.hist(nir[:,:,2].ravel(),bins=255)
plt.axis("off")
```
